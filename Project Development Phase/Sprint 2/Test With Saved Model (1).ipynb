```python
#Importing the Keras libraries and packages
from tensorflow.keras.models import load_model
model = load_model(r'F:\semester 7\ibm\SI-GuidedProject-29143-1667502003-main\SI-GuidedProject-29143-1667502003-main\Project Development Phase\Sprint 2\project.h5')

```

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test)=mnist.load_data () #splitting the mnist data into train and test
```

```python
from sklearn.metrics import confusion_matrix

predicted = model.predict(x_test)
```

```python
predicted
```

```python
import numpy as np
predicted_label = np.argmax(predicted, axis=1)
```

```python
y_test
```

Confusion Matrix

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score 
import matplotlib.pyplot as plt 
conf_matrix = confusion_matrix(y_test, predicted_label) # 
# Print the confusion matrix using Matplotlib # 
fig, ax = plt.subplots(figsize=(5, 5)) 
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3) 
for i in range(conf_matrix.shape[0]): 
    for j in range(conf_matrix.shape[1]): 
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large') 
plt.xlabel('Predictions', fontsize=18) 
plt.ylabel('Actuals', fontsize=18) 
plt.title('Confusion Matrix', fontsize=18) 
plt.show()
```

```python
print('Precision: %.3f' % precision_score(y_test, predicted_label, pos_label='positive', average='micro'))
```

```python
print('Recall: %.3f' % recall_score(y_test, predicted_label, pos_label='positive', average='micro'))
```

```python
print('Accuracy: %.3f' % accuracy_score(y_test, predicted_label))
```

```python
print('F1 Score: %.3f' % f1_score(y_test, predicted_label, pos_label='positive', average='micro'))
```
